# Plots

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import scipy.stats as sps
```

```python
from math import log10, floor
def round_sig(x, sig=3):
    return round(x, sig-int(floor(log10(abs(x))))-1)
```

```python
EPRScnak = np.array(pd.read_csv('EPRhmm.csv'))
EPR_true = np.array(pd.read_csv('EPRTrue.txt',header=None)[0])
henrici = np.array(pd.read_csv('henrici.csv'))
nhenrici = np.array(pd.read_csv('nhenrici.csv'))
td = np.array(pd.read_csv('td.csv'))
irrecip = np.array(pd.read_csv('irrecip.csv'))



fig, axs = plt.subplots(2,3)
fig.set_size_inches(10,6)

axs[0,0].plot(EPR_true,np.mean(EPRScnak,0),color = 'orangered')
stdup = np.mean(EPRScnak,0)+ np.std(EPRScnak,0)
stddown = np.mean(EPRScnak,0)- np.std(EPRScnak,0)
axs[0,0].fill_between(EPR_true, stddown, stdup, color="orangered", alpha=0.2)
axs[0,0].set_xlabel(r'$\Phi^{GT}$')
axs[0,0].set_ylabel(r'$\Phi^{HMM}$')
rho = sps.pearsonr(EPR_true,np.mean(EPRScnak,0))
axs[0,0].text(.05, 1, r'$\rho = $'+ str(round_sig(rho.statistic)), transform=axs[0,0].transAxes)

axs[0,1].plot(EPR_true,np.mean(henrici,0),color = 'orchid')
stdup = np.mean(henrici,0)+ np.std(henrici,0)
stddown = np.mean(henrici,0)- np.std(henrici,0)
axs[0,1].fill_between(EPR_true, stddown, stdup, color="orchid", alpha=0.2)
axs[0,1].set_xlabel(r'$\Phi^{GT}$')
axs[0,1].set_ylabel(r'$d_H$')
rho = sps.pearsonr(EPR_true,np.mean(henrici,0))
axs[0,1].text(.05, 1, r'$\rho = $'+ str(round_sig(rho.statistic)), transform=axs[0,1].transAxes)


axs[0,2].plot(EPR_true,np.mean(nhenrici,0),color = 'darkslategrey')
stdup = np.mean(nhenrici,0)+ np.std(nhenrici,0)
stddown = np.mean(nhenrici,0)- np.std(nhenrici,0)
axs[0,2].fill_between(EPR_true, stddown, stdup, color="darkslategrey", alpha=0.2)
axs[0,2].set_xlabel(r'$\Phi^{GT}$')
axs[0,2].set_ylabel(r'$\hat{d}_H$')
rho = sps.pearsonr(EPR_true,np.mean(nhenrici,0))
axs[0,2].text(.05, 1, r'$\rho = $'+ str(round_sig(rho.statistic)), transform=axs[0,2].transAxes)


axs[1,0].plot(EPR_true,np.mean(td,0),color = 'forestgreen')
stdup = np.mean(td,0)+ np.std(td,0)
stddown = np.mean(td,0)- np.std(td,0)
axs[1,0].fill_between(EPR_true, stddown, stdup, color="forestgreen", alpha=0.2)
axs[1,0].set_xlabel(r'$\Phi^{GT}$')
axs[1,0].set_ylabel(r'$\sqrt{1-F_0}$')
rho = sps.pearsonr(EPR_true,np.mean(td,0))
axs[1,0].text(.05, 1, r'$\rho = $'+ str(round_sig(rho.statistic)), transform=axs[1,0].transAxes)


axs[1,1].plot(EPR_true,np.mean(irrecip,0),color = 'midnightblue')
stdup = np.mean(irrecip,0)+ np.std(irrecip,0)
stddown = np.mean(irrecip,0)- np.std(irrecip,0)
axs[1,1].fill_between(EPR_true, stddown, stdup, color="midnightblue", alpha=0.2)
axs[1,1].set_xlabel(r'$\Phi^{GT}$')
axs[1,1].set_ylabel(r'$1-r$')
rho = sps.pearsonr(EPR_true,np.mean(irrecip,0))
axs[1,1].text(.05, 1, r'$\rho = $'+ str(round_sig(rho.statistic)), transform=axs[1,1].transAxes)

axs[1,2].axis('off')

sns.despine()
plt.tight_layout()
plt.show()
```
